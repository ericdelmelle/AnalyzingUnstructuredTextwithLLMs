# Google Search
## No key needed

- Here we show how we can get data from google searc, without the need for a key
- We will identify coffee shops near a zipcode

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import bs4

location = "19129"
query = "coffee " + location
print("https://www.google.com/search?q=" + query)

webpage = requests.get("https://www.google.com/search?q=" + query)
print(webpage)
```

raw html is returned

```python
#Take the webpage variable and fetch the content using BeautifulSoup
soup = BeautifulSoup(webpage.content, "html.parser")
print(soup)
```

Let's make this a little more pretty!

```python
heading_object=soup.find_all( 'h3' )
# Iterate through the object
# and print it as a string.
for info in heading_object:
    print(info.getText())
    print("------")

# Pulling HTTP data from internet
soupB = bs4.BeautifulSoup( soup.text
                         , "html.parser" )
print(soupB)
```
