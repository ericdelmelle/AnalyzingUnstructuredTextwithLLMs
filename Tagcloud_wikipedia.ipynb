# WORDCLOUDS **TAGCLOUDS**

- A tag cloud, also called a word cloud or weighted list in visual design, visually represents text data, often highlighting keywords or metadata on websites or in free-form text. 

- Typically, each tag is a single word, and its significance is represented by font size or color. When used for website navigation, tags are usually linked to related items.


```python
#Import package
!pip install wikipedia

import wikipedia
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS# Generate word cloud

#check function wordcloud.wordcloud here:
#https://amueller.github.io/word_cloud/generated/wordcloud.WordCloud.html

def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(10, 5))
    # Display image
    plt.imshow(wordcloud)
    # No axis details
    plt.axis("off");
```

```python
# Specify the title of the Wikipedia page
wiki = wikipedia.page('Geographic_information_system')# Extract the plain text content of the page
text = wiki.content
#text =  open("summary.txt", "r").read()


## Clean text
text = re.sub(r'==.*?==+', '', text)

wordcloud = WordCloud(width= 3000, height = 2000, random_state=1,
                      background_color='salmon', colormap='Pastel1',
                      max_words = 100,
                      collocations=False, stopwords = STOPWORDS).generate(text)# Plot
plot_cloud(wordcloud)
```
