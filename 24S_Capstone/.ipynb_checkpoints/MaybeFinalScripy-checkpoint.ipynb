```python
import googlemaps as gmaps

#!pip install pandas
import pandas as pd

#!pip install geopandas
import geopandas as gpd

#!pip install seaborn
import seaborn as sns

#!pip install matplotlib
from matplotlib import pyplot as plt


import requests
```

```python
park_df = gpd.read_file('PPR_Properties.geojson')

google_reviews_df = pd.read_csv('park_google_reviews.csv')
```

```python
## filter out non_park values

exclude_values = [
    #"BREEZEWAY_GREENWAY",
    "CONSERVATION_UNDEVELOPED",
    "BOAT_SITE",
    #"FARM",
    "EVENT_VENUE",
    "GOLF_FACILITY", 
    "PPR_OPERATIONS_FACILITY",
    "RECREATION_CENTER",
    "OLDER_ADULT_CENTER",
    "POOL_SITE"
]

park_df = park_df[~park_df['PROPERTY_CLASSIFICATION'].isin(exclude_values)]
```

```python
## rename some parks i recognized
rename_values = {
    "JFK Plaza (Love Park)": "John F Kennedy Plaza (LOVE Park)",
    "Clark Park": "Clarence H Clark Park",
    "Borski Park": "Robert A Borski"
}

park_df['PUBLIC_NAME'] = park_df['PUBLIC_NAME'].replace(rename_values)

#and make sure uniqueness
park_df = park_df.drop_duplicates(subset='PUBLIC_NAME')
```

```python
## Join review data into park info by name

reviews_df = park_df.merge(google_reviews_df, left_on='PUBLIC_NAME', right_on='park_name', how='left')

reviews_df.head()

reviews_df.to_csv('sample_reviews.csv', index=False)
```

# Sentiment Score

```python
# apply the model calculating the score

#!pip install tensorflow
#!pip3 install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
from transformers import pipeline

# Load the sentiment analysis pipeline
pipe = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Make sure all entries in the 'text' column are strings
reviews_df['text'] = reviews_df['text'].astype(str)

# Apply the pipeline to each review
reviews_df['review_predictions'] = reviews_df['text'].apply(lambda x: pipe(x, truncation=True))

# Extract the label and score from the predictions
reviews_df['review_label'] = reviews_df['review_predictions'].apply(lambda x: x[0]['label'])
reviews_df['review_score'] = reviews_df['review_predictions'].apply(lambda x: x[0]['score'])
```

```python
# Transform Negative and Positive Sentiment

reviews_df['sentiment'] = reviews_df.apply(lambda x: x['review_score'] if x['review_label'] == 'POSITIVE' else (-x['review_score'] if x['review_label'] == 'NEGATIVE' else 0), axis=1)


reviews_df.head()
```

```python
## Only Monthly Sentiment Shift with Confident Interval
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Calculate mean and standard deviation of the sentiment scores for each month
monthly_stats = reviews_df.groupby(reviews_df['time'].dt.strftime('%b')).agg({
    'sentiment': ['mean', 'std', 'count']
}).reset_index()

# Rename columns for easier access
monthly_stats.columns = ['month', 'mean_sentiment', 'std_sentiment', 'count']

# Sort the months in calendar order
months_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_stats['month'] = pd.Categorical(monthly_stats['month'], categories=months_order, ordered=True)
monthly_stats = monthly_stats.sort_values('month')

# Calculate the standard error of the mean (SEM) for the sentiment score
monthly_stats['sem_sentiment'] = monthly_stats['std_sentiment'] / np.sqrt(monthly_stats['count'])

# Assuming a normal distribution, we can use 1.96 * SEM for the 95% confidence interval
# This is an approximation and for precise results, other methods like bootstrapping could be used
monthly_stats['ci95_high'] = monthly_stats['mean_sentiment'] + 1.96 * monthly_stats['sem_sentiment']
monthly_stats['ci95_low'] = monthly_stats['mean_sentiment'] - 1.96 * monthly_stats['sem_sentiment']

# Plot the mean sentiment score for each month
plt.figure(figsize=(10, 5))
plt.plot(monthly_stats['month'], monthly_stats['mean_sentiment'], marker='o', color='tab:blue', label='Average Sentiment Score')

# Add a shaded area for the confidence interval
plt.fill_between(monthly_stats['month'], monthly_stats['ci95_low'], monthly_stats['ci95_high'], color='tab:blue', alpha=0.2)

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Average Sentiment Score')
plt.title('Average Sentiment Score with 95% Confidence Interval by Month')
plt.grid(True)
plt.tight_layout()
plt.show()


## remove the boader and grid lines, and data markers, the axes label
```

```python
from collections import Counter
import string
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords

def content_analysis(df, text_column, emotion_label, words_to_drop):
    # Convert text to lowercase, remove punctuation, and split into words
    words = df[text_column].str.lower().str.replace(f'[{string.punctuation}]', '', regex=True).str.split()
    words_flat = [word for sublist in words for word in sublist]
    
    # Remove stopwords and unwanted words
    stop_words = set(stopwords.words('english'))
    words_no_stop = [word for word in words_flat if word not in stop_words and word not in words_to_drop]
    
    # Count the frequency of each word
    word_counts = Counter(words_no_stop)
    
    # Convert to a DataFrame
    words_df = pd.DataFrame(word_counts.items(), columns=['word', 'count']).sort_values(by='count', ascending=False)
    
    # Generate pie chart with a title and a color palette
    words_df.head(10).set_index('word').plot(
        kind='pie', 
        y='count', 
        autopct='%1.1f%%', 
        startangle=90, 
        colormap='viridis', # Set the color palette
        figsize=(8, 8),
        legend=False
    )
    plt.title(f'Word Frequency for {emotion_label.capitalize()} Reviews')
    plt.ylabel('')  # Hide the y-label
    plt.show()
    
    return words_df
```

```python
words_to_drop = ['park', 'nan','s','*','â€™']  # Add more words as needed

joy_word_counts = content_analysis(joy_df, 'text', 'joy', words_to_drop)
fear_word_counts = content_analysis(fear_df, 'text', 'fear', words_to_drop)
sadness_word_counts = content_analysis(sadness_df, 'text', 'sadness', words_to_drop)
anger_word_counts = content_analysis(anger_df, 'text', 'anger', words_to_drop)
love_word_counts = content_analysis(love_df, 'text', 'love', words_to_drop)

## how to merge single and pulural??!! or do it manually 
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.corpus.reader.xmldocs import XMLCorpusReader
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def verb_analysis(df, text_column, emotion_label, words_to_drop):
    # Tokenize the text and tag part of speech
    df.loc[:, 'tokens'] = df.loc[:, text_column].apply(nltk.word_tokenize)
    df['pos_tags'] = df['tokens'].apply(nltk.pos_tag)
    
    # Filter out tokens to get verbs
    df['verbs'] = df['pos_tags'].apply(lambda x: [word for word, pos in x if pos.startswith('VB')])
    
    # Flatten the list of verbs and remove stopwords and words to drop
    all_verbs_flat = [verb for sublist in df['verbs'] for verb in sublist]
    stop_words = set(stopwords.words('english'))
    verbs_filtered = [verb for verb in all_verbs_flat if verb.lower() not in stop_words and verb.lower() not in words_to_drop]
    
    # Define unwanted tokens
    unwanted_tokens = {"'s", "'","'ve", "'", "â€™"}

    # Filter them out
    verbs_filtered = [verb for verb in all_verbs_flat if verb.lower() not in stop_words and verb.lower() not in words_to_drop and verb not in unwanted_tokens]
    
    # Count the frequency of each verb
    verb_counts = Counter(verbs_filtered)
    
    # Convert to a DataFrame
    verbs_df = pd.DataFrame(verb_counts.items(), columns=['verb', 'count']).sort_values(by='count', ascending=False)
    
    # Generate pie chart with a title and a color palette
    verbs_df.head(10).set_index('verb').plot(
        kind='pie',
        y='count',
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.viridis(np.linspace(0, 1, 10)), # Use a colormap for variety
        figsize=(8, 8),
        legend=False
    )
    plt.title(f'Verb Frequency for {emotion_label.capitalize()} Reviews')
    plt.ylabel('')  # Hide the y-label
    plt.show()
    
    return verbs_df
```

```python
joy_verb_counts = verb_analysis(joy_df, 'text', 'joy', words_to_drop)
fear_verb_counts = verb_analysis(fear_df, 'text', 'fear', words_to_drop)
sadness_verb_counts = verb_analysis(sadness_df, 'text', 'sadness', words_to_drop)
anger_verb_counts = verb_analysis(anger_df, 'text', 'anger', words_to_drop)
love_verb_counts = verb_analysis(love_df, 'text', 'love', words_to_drop)
```

```python
## noun
import nltk
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

# Download necessary NLTK resources
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('wordnet')  # For the lemmatizer

# Lemmatizer will help in converting plurals to singular
from nltk.stem import WordNetLemmatizer

def noun_analysis(df, text_column, emotion_label, words_to_drop):
    # Initialize the lemmatizer
    lemmatizer = WordNetLemmatizer()
    
    # Tokenize the text and tag part of speech
    df['tokens'] = df[text_column].apply(nltk.word_tokenize)
    df['pos_tags'] = df['tokens'].apply(nltk.pos_tag)
    
    # Filter out tokens to get nouns (common and proper nouns)
    df['nouns'] = df['pos_tags'].apply(lambda x: [word for word, pos in x if pos in ('NN', 'NNS', 'NNP', 'NNPS')])
    
    # Flatten the list of nouns, lemmatize them, and remove stopwords and words to drop
    all_nouns_flat = [lemmatizer.lemmatize(noun.lower()) for sublist in df['nouns'] for noun in sublist]
    stop_words = set(stopwords.words('english'))
    nouns_filtered = [noun for noun in all_nouns_flat if noun not in stop_words and noun not in words_to_drop]
    
    # Count the frequency of each noun
    noun_counts = Counter(nouns_filtered)
    
    # Convert to a DataFrame
    nouns_df = pd.DataFrame(noun_counts.items(), columns=['noun', 'count']).sort_values(by='count', ascending=False)
    
    # Generate pie chart with a title and a color palette
    nouns_df.head(10).set_index('noun').plot(
        kind='pie',
        y='count',
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.viridis(np.linspace(0, 1, 10)), # Use a colormap for variety
        figsize=(8, 8),
        legend=False
    )
    plt.title(f'Noun Frequency for {emotion_label.capitalize()} Reviews')
    plt.ylabel('')  # Hide the y-label
    plt.show()
    
    return nouns_df
```

```python
joy_noun_counts = noun_analysis(joy_df, 'text', 'joy', words_to_drop)
fear_noun_counts = noun_analysis(fear_df, 'text', 'fear', words_to_drop)
sadness_noun_counts = noun_analysis(sadness_df, 'text', 'sadness', words_to_drop)
anger_noun_counts = noun_analysis(anger_df, 'text', 'anger', words_to_drop)
love_noun_counts = noun_analysis(love_df, 'text', 'love', words_to_drop)
```

```python
## try LDA
!pip install gensim
from gensim import corpora
from gensim.models.ldamodel import LdaModel
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import pandas as pd

nltk.download('stopwords')

def perform_lda_topic_modeling(df, text_column, num_topics=5, num_words=4):
    tokenizer = RegexpTokenizer(r'\w+')
    p_stemmer = PorterStemmer()

    # Create a list of stopwords
    stop_words = set(stopwords.words('english'))

    # List for tokenized documents
    texts = []

    # Loop through document list
    for i in df[text_column]:
        # Lowercase and tokenize document
        raw = i.lower()
        tokens = tokenizer.tokenize(raw)
        # Remove stopwords
        stopped_tokens = [i for i in tokens if not i in stop_words]
        # Stem tokens
        stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
        # Add tokens to list
        texts.append(stemmed_tokens)

    # Turn our tokenized documents into a id <-> term dictionary
    dictionary = corpora.Dictionary(texts)

    # Convert tokenized documents into a document-term matrix
    corpus = [dictionary.doc2bow(text) for text in texts]

    # Generate LDA model
    ldamodel = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20)

    # Print topics
    topics = ldamodel.print_topics(num_words=num_words)
    for topic in topics:
        print(topic)

    return topics
```

```python
from collections import Counter, OrderedDict
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
from gensim import corpora
from gensim.models.ldamodel import LdaModel
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import numpy as np

# Make sure to download stopwords
nltk.download('stopwords')

def perform_lda_topic_modeling(df, text_column, num_topics=5, num_words=4, additional_stopwords=set()):
    tokenizer = RegexpTokenizer(r'\w+')
    p_stemmer = PorterStemmer()

    # Extend the list of stopwords
    stop_words = set(stopwords.words('english')).union(additional_stopwords)
    
    # Prepare texts
    texts = []  # This will hold the tokenized documents

    # Tokenize each document
    for i in df[text_column]:
        # Tokenize
        raw = i.lower()
        tokens = tokenizer.tokenize(raw)
        
        # Remove stopwords and stem
        stopped_tokens = [i for i in tokens if not i in stop_words]
        stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
        
        # Add to list
        texts.append(stemmed_tokens)

    # Create a dictionary and corpus
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]

    # Build LDA model
    lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20)

    return lda_model, corpus, dictionary, texts  # Return the texts for visualization

def lda_visualization(ldamodel, corpus, texts, num_topics=5, num_words=3):
    # Ensure the texts are not empty
    if not texts:
        raise ValueError("The text data is empty. Cannot visualize topics.")
    
    # Extract topic weights and words
    topics = ldamodel.show_topics(num_topics=num_topics, num_words=num_words, formatted=False)
    data_flat = [w for w_list in texts for w in w_list]  # Flatten the list of texts
    counter = Counter(data_flat)

    # Aggregate and present data for topics and respective weights
    # Make sure the keys of the topics_dict match the DataFrame construction
    topics_dict = {str(i): OrderedDict(ldamodel.show_topic(i, 25)) for i in range(num_topics)}
    df = pd.DataFrame(topics_dict)

    # If DataFrame is empty after this point, there's likely an issue with the model output
    if df.empty:
        raise ValueError("The topic modeling did not produce any topics.")

    df = df.fillna(0).T.sort_values(by=str(0), ascending=False)  # Use str(0) to access the column
    df = df.head(num_topics)

    # Prepare the topic weights for plotting
    topic_weights = df.sum(axis=1).sort_values(ascending=False)  # Corrected to use sum of each topic's words

    # Plot the data
    fig, ax = plt.subplots(1, 2, figsize=(10, 4), dpi=120, gridspec_kw={'width_ratios': [3, 2]})

    # Topic Weights
    ax[0].barh(range(len(topic_weights)), topic_weights)
    ax[0].set_yticks(range(len(topic_weights)))
    ax[0].set_yticklabels(['Topic {}'.format(i) for i in range(len(topic_weights))])
    ax[0].invert_yaxis()
    ax[0].set_title('Creation of topics')

    # Topics allocation to documents - ensure doc_topic_dist is computed correctly
    doc_topic_dist = [max(prob, key=lambda y:y[1])[0] for prob in ldamodel.get_document_topics(corpus)]
    ax[1].hist(doc_topic_dist, bins=range(num_topics+1), align='left', rwidth=0.5)
    ax[1].set_xticks(range(num_topics))
    ax[1].set_xticklabels(['Topic {}'.format(i) for i in range(num_topics)])
    ax[1].set_title('Topics allocation to documents')

    plt.tight_layout()
    plt.show()
```

```python
def lda_modeling_and_visualization(df, text_column, additional_stopwords=set(), num_topics=5):
    # Preprocessing
    tokenizer = RegexpTokenizer(r'\w+')
    p_stemmer = PorterStemmer()
    stop_words = set(stopwords.words('english')).union(additional_stopwords)
    
    # Tokenize, stopword removal, and stemming
    texts = [
        [p_stemmer.stem(word) for word in tokenizer.tokenize(document.lower()) if word not in stop_words]
        for document in df[text_column]
    ]
    
    # Create the Dictionary and Corpus
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    # LDA model
    lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20)
    
    # Topic visualization
    topic_weights = []
    for i in range(num_topics):
        topic_weights.extend(lda_model.show_topic(i, topn=5))
    
    # Get the topic numbers for each weight
    topic_numbers = [topic[0] for topic in topic_weights]
    # Get the word weights for each topic
    word_weights = [topic[1] for topic in topic_weights]
    # Generate a DataFrame from the topic weights
    df_topic_weights = pd.DataFrame(list(zip(topic_numbers, word_weights)), columns=['Topic', 'Weight'])
    df_topic_weights['Topic'] += 1  # Start topic numbering at 1
    df_topic_weights['Word'] = [topic[0] for topic in topic_weights]

    # Plot the topic distribution
    plt.figure(figsize=(10, 5))
    df_topic_weights.groupby('Topic').agg({'Weight': 'sum'}).sort_values('Weight', ascending=False).plot(kind='bar')
    plt.title('Topic Distribution')
    plt.ylabel('Weight')
    plt.xlabel('Topic')
    plt.xticks(rotation=0)
    plt.show()
    
    # Topic allocation to documents
    topic_allocation = np.zeros((df.shape[0], num_topics))
    for i, row in enumerate(corpus):
        topic_distribution = lda_model.get_document_topics(row)
        for topic_num, prob in topic_distribution:
            topic_allocation[i, topic_num] = prob
    
    # Create a DataFrame to display the topic allocation to documents
    df_topic_allocation = pd.DataFrame(topic_allocation, columns=[f'Topic {i+1}' for i in range(num_topics)])
    print(df_topic_allocation)

    
additional_stopwords = {'philadelphia', 'park','nan'}
```

```python
## or BERT
!pip install --upgrade transformers
!pip install --upgrade sentence-transformers
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import pandas as pd

def perform_bert_clustering(df, text_column, num_clusters=5):
    # Load the BERT model
    model = SentenceTransformer('bert-base-nli-mean-tokens')

    # Encode the text data to get the embeddings
    document_embeddings = model.encode(df[text_column].tolist())

    # Perform kmeans clustering
    clustering_model = KMeans(n_clusters=num_clusters)
    clustering_model.fit(document_embeddings)
    cluster_assignment = clustering_model.labels_

    # Add cluster assignment to dataframe
    df['Topic'] = cluster_assignment

    return df[['text_column', 'Topic']]
```
